### Problem 1: Root finding

**(a) & (b)** Newton's Method to find roots of equations using the following formula:

$$x_{i+1} = x_{i} - \frac{f(x_{i})}{f'(x_{i})}$$

Start at $x=0$, until $|x_{i} - x_{i-1}| \leqslant 0.0001$

<br>

- $f(x) = 20x^3 − 50x^2 + 3x + 20$
<br>$f'(x) = 60x^2 - 100x + 3$


```python
from math import exp, log

# Function to find roots of
def f(x):
    return 20*x**3 - 50*x**2 + 3*x + 20 
    
# Derivation of function
def df(x):
    return 60*x**2 - 100*x + 3 

def main():
    # Given parameters
    initial = 0
    tolerance = 0.0001

    # Initialising iteration counter
    value = 0

    # Using to compute |x_{i} − x_{i−1}|
    next = initial - (f(initial) / df(initial))
    diff = abs(next - initial)

    # Printing initial two lines of output
    print("Iteration    " + "Xi         " + "f(Xi)")
    print("0           ", '{:.5f}'.format(initial), "  ", '{:.5f}'.format(f(initial))) #Formatting floats to 5 decimal places

    # Main algorithm for newton's method
    while not (diff <= tolerance):
        value += 1
        next = initial - (f(initial) / df(initial))
        diff = abs(next - initial)
        initial = next # Swapping values
        if not (diff <= tolerance):
            print(value, "          ", '{:.5f}'.format(next), "  ", '{:.5f}'.format(f(next)))
main()
```

**(c)** Using bisection method to find roots of:

$$f(x) = 20x^3 − 50x^2 + 3x + 20$$

Given interval -> {(a, b) = (0, 1)}

```python
# Function to find roots of
def f(x):
    return (20*x**3 - 50*x**2 + 3*x + 20) 

# Given values of a and b, where function f is defined.
a = 0.0
b = 1.0

# Given tolerance
tolerance = 0.0001

# Printing initial line of output
print("Iteration", "     ", "a", "         ", "b", "        ", "f(c)")
 
# Bisection method function
def bisection(a, b, f, tolerance): 
    iter = 0 # Iteration counter
    max = 50 # Ensuring algorithm doesn't run too long
    while iter < max:
        midpt = (a + b)/2 # Calculating mid-point
        func_mid = f(midpt)
        result = print(iter, "          ", '{:.5f}'.format(a), "   ", '{:.5f}'.format(b), "   ", '{:.5f}'.format(func_mid))
        if (abs(b - a) < tolerance):
            break
        iter += 1
        if f(a) * func_mid > 0: # Checking for signs (+ or -)
            a = midpt
        else:
            b = midpt
    return result

bisection(a, b, f, tolerance)
```

### Problem 2: SVD based data compression

#### Reading and Displaying PGM file
**(a)** Including a 120px x 100px PNG image, converted to grayscale and PGM file. <br>
Performing SVD on image for compression & illustrating results by components $A = UDV^{T}$

```python
from matplotlib import pyplot
import re
import numpy
import numpy.linalg
import matplotlib.pyplot as plt


def read_pgm(filename, byteorder='>'):
    """Return image data from a raw PGM file as numpy array.

    Format specification: http://netpbm.sourceforge.net/doc/pgm.html

    """
    with open(filename, 'rb') as f:
        buffer = f.read()
    try:
        header, width, height, maxval = re.search(
            b"(^P5\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n])*"
            b"(\d+)\s(?:\s*#.*[\r\n]\s)*)", buffer).groups()
    except AttributeError:
        raise ValueError("Not a raw PGM file: '%s'" % filename)
    return numpy.frombuffer(buffer,
                            dtype='u1' if int(maxval) < 256 else byteorder+'u2',
                            count=int(width)*int(height),
                            offset=len(header)
                            ).reshape((int(height), int(width)))


def display_pgm(array):
    pyplot.imshow(array, pyplot.cm.gray)
    pyplot.axis('off')
    pyplot.show()

image = read_pgm("sabaoon.pgm")

# Displaying original image
print("\nA =")
display_pgm(image)

# Performing the SVD, A = UDV^T of the facial image:
U, D, V = numpy.linalg.svd(image)
D = numpy.diag(D)
Vt = numpy.transpose(V)

# Finding max & min values for each matrix/array
Umax = U.max()
Umin = U.min()
Dmax = D.max()
Dmin = D.min()
Vmax = V.max()
Vmin = V.min()

# Finding number of rows and columns for each matrix
Uheight, Ulength = U.shape
Dheight, Dlength = D.shape
Vheight, Vlength = V.shape


# For visualisation, linearly mapping values of the matrix elements Uij , Dii, and Vij to the range [0, 255].

# Function to rescale matrix values
def rescale(matrix, height, length, minimum, maximum):
    new_matrix = numpy.empty([height, length]) # Creating a new matrix with the same dimensions, so the original isn't changed
    for i in range(height):
        for j in range(length):
            new_matrix[i, j] = 255 * ((matrix[i, j] - minimum) / (maximum - minimum)) # Given rescaling formula
    return new_matrix

U_rescaled = rescale(U, Uheight, Ulength, Umin, Umax)
D_rescaled = rescale(D, Dheight, Dlength, Dmin, Dmax)
V_rescaled = rescale(Vt, Vheight, Vlength, Vmin, Vmax)

# Displaying rescaled U, D, Vt components
print("U =")
display_pgm(U_rescaled)
print("D =")
display_pgm(D_rescaled)
print("Vt =")
display_pgm(V_rescaled)
```

**(b)** Approximating A with ρ singular values and columns of U and V for 1 ≤ ρ ≤ 100 and illustrating
the obtained results.

```python
# To visualise every approximate matrix,
# values that are less than 0 are set to 0 and the values larger than 255 are set to 255.
def resetting(matrix):
    height, length = matrix.shape
    for i in range(height):
        for j in range(length):
            if matrix[i, j] < 0:
                matrix[i, j] = 0
            if matrix[i, j] > 255:
                matrix[i, j] = 255
    return matrix 

# Approximation code here:
for i in range(1, 101):
    approximation = numpy.matrix(U[:, :i]) * D[:i,:i] * numpy.matrix(V[:i, :])
    resetting(approximation)
    pyplot.title(i)
    display_pgm(approximation)
```

**(c)** Finding the absolute approximation errors and the level of compression with respect to the original image for different values of ρ.

```python
# Finding the absolute error
def abs_error(U, D, V, A, i):
    approx = numpy.matrix(U[:, :i]) * D[:i,:i] * numpy.matrix(V[:i, :])
    return abs(A - approx)

# Outputting the main heading
print("ρ      ", "Max error     ", "Mean error     ", "Compression %")

# Image Dimensions
height = 120
width = 100

for p in range(1, 6):
    absoluteError = abs_error(U, D, V, image, p)
    max_val = absoluteError.max() # Calculating the maximum error from the absolute error array
    mean_val = absoluteError.mean() # Calculating the mean error from the absolute error array
    compression = (1 - (p * (1 + width + height) / 12000)) * 100 # Applying the compression formula (1 - k/12000)
    print(p,"      ",'{:.2f}'.format(max_val),"         ",'{:.2f}'.format(mean_val),"         ",'{:.1f}'.format(compression))
    
for p in range(10, 100, 10):
    absoluteError = abs_error(U, D, V, image, p)
    max_val = absoluteError.max()
    mean_val = absoluteError.mean()
    compression = (1 - (p * (1 + width + height) / 12000)) * 100
    print(p,"      ",'{:.2f}'.format(max_val),"         ",'{:.2f}'.format(mean_val),"         ",'{:.1f}'.format(compression))
```
